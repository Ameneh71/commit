```python
import pandas as pd
df = pd.read_stata('election_campaigns_2023_corrected.dta')
for c in df:
    if df[c].dtype == 'float':
        try:
            df[c] = df[c].astype(pd.Int64Dtype())
        except:
            print(c)
df
```

```python
cols = [
  'country_year',
  # 'issue_cat',
  # 'issue_subcat',
  'party_id',
  'issue_subcat',
]
total = df[df['type_cs'] == 'actor-issue'].groupby(cols)['direction'].sum()
```

```python
total_nonzero = total[total != 0].sort_values()
total_nonzero.to_csv('election_campaigns_2023_corrected.total_direction.csv', 
                     index=None)
total_nonzero
```

```python
party_info = [
    'country_year',
    'year_election',
    'country',
    'iso2code',
    'party_id',
    'party_name',
    'party_name_short',
    'coalition',
    'vote_share',
    'subject_organization',
    # 'seats',
    # 'seats_total',
]
parties = df[party_info].drop_duplicates()
parties
```

```python
def election_result(df):
    vote_share_per_year = df.set_index('country_year').sort_index()['vote_share']
    return vote_share_per_year.diff() / vote_share_per_year

rel_vote_share_change = parties.groupby('party_id').apply(election_result).dropna()
```

```python
import rdflib, re, tqdm

rel_vote_share_change_cats = rvscc = pd.qcut(rel_vote_share_change, 5, 
    labels=["bigloss", "loss", "nochange", "win", "bigwin"]).reset_index()

direction_cats = dc = pd.qcut(total_nonzero, 5, 
    labels=["vneg", "neg", "neut", "pos", "vpos"]).reset_index()

parties

base = rdflib.Namespace('http://jieyingchenchen.github.io/polsci-gnn/')

a = rdflib.RDF.type
for country_year in tqdm.tqdm(set(parties['country_year'])):
    g = rdflib.Graph()
    g.bind(None, base)
    for p in ['election', 'result', 'party','result', 'issue', 'stance', 'strength']:
        g.bind(p, base[p+'/'])
    election = base[f'election/{country_year}']
    g.add( (election, a, base['Election']) )

    for i, row in rvscc[rvscc['country_year'] == country_year].iterrows():
        s = base[f'result/{i}']
        g.add( (s, a, base['ElectionResult']) )
        g.add( (s, base['election'], election) )
        g.add( (s, base['party'], base[f'party/{row.pop("party_id")}']) )
        g.add( (s, base['result'], base[f'result/{row.pop("vote_share")}'] ) )
    
    for i, row in dc[dc['country_year'] == country_year].iterrows():
        s = base[f'stance/{i}']
        issue = '_'.join(re.findall('\w+', row.pop("issue_subcat")))
        g.add( (s, a, base['Stance']) )
        g.add( (s, base['election'], election) )
        g.add( (s, base['party'], base[f'party/{row.pop("party_id")}']) )
        g.add( (s, base['issue_subcat'], base[f'issue/{issue}'] ) )
        g.add( (s, base['strength'], base[f'strength/{row.pop("direction")}'] ) )

    for i, row in dc[dc['country_year'] == country_year].iterrows():
        s = base[f'party/{row.pop("party_id")}']
        g.add( (s, a, base['Party']) )
        # for k,v in row.items():
        #     g.add( (s, base[k], base[v]) )


    g.serialize(f'graphs/{country_year}.ttl')
```
